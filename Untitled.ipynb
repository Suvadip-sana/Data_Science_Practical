```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def prediction(x1, x2, w1, w2):
    s = x1*w1 + x2*w2
    if s>= 0.5:
        return 1
    else:
        return 0
    
```

```python
def error(x1, x2, w1, w2, y):
    return y - prediction(x1, x2, w1, w2)
```

```python
def lossFunction(x1, x2, w1, w2, y):
    sum = 0
    for i in range(len(x1)):
        sum += error(x1[i], x2[i], w1, w2, y[i])
    return sum
```

```python
def traningFunc(x1, x2, y, lr, initialW1, initialW2):
    w1 = initialW1
    w2 = initialW2
    lossList = []
    for epoch in (range(100)):
        loss = lossFunction(x1, x2, w1, w2, y)
        lossList.append(loss)
        for i in range(len(x1)):
            w1 = w1 + lr * error(x1[i], x2[i], w1, w2, y[i]) * x1[i]
            w2 = w2 + lr * error(x1[i], x2[i], w1, w2, y[i]) * x2[i]

    return w1, w2, lossList


        
```

```python
def linePoint(x1, w1, w2, th):
    x2 = []
    for i in range(len(x1)):
        x2.append((0.5-x1[i]*w1)/w2)
    return x2
```

# OR Function

```python
x1 = [0, 0, 1, 1]
x2 = [0, 1, 0, 1]
y = [0, 1, 1, 1]
```

```python
w1, w2, losses = traningFunc(x1, x2, y, 0.01, 0.7, 0.2)
```

```python
print(w1, w2)
```

```python
print(losses)
```

```python
epochs = [x for x in range(len(losses))]
```

```python
plt.plot(epochs, losses, color='b', marker='^')
plt.xlabel('epochs')
plt.ylabel('losses')
plt.show()
```

```python
x2Predict = linePoint(x1, w1, w2, 0.5)
```

```python
plt.scatter(x1[0], x2[0], color='b', marker='d')
plt.scatter(x1[1:], x2[1:], color='r', marker='^')
plt.plot(x1, x2Predict, color = 'g')
plt.show()
```

# KNN Classification using Pima Dataset

### What is the K-Nearest Neighbors (K-NN) Algorithm?
    The K-Nearest Neighbors (K-NN) algorithm is a simple, non-parametric, and instance-based supervised learning method used for classification and regression. In K-NN, the classification of a data point is determined based on the majority class of its nearest neighbors. The algorithm makes no assumptions about the underlying data distribution, which makes it highly versatile. However, its performance heavily depends on the value of k (the number of neighbors) and the choice of the distance metric. For classification tasks, K-NN assigns the most common class among the k-nearest data points.

### Steps of the K-NN Algorithm:
#### 1. Load the Data: 
Collect and preprocess the dataset, ensuring features are scaled properly to avoid bias from varying feature magnitudes.

#### 2. Choose the Value of k:
Select the number of neighbors (k) to consider. A smaller k results in higher sensitivity to noise, while a larger k can smooth out classifications.

#### 3. Compute the Distance:
For a given test point, compute the distance to all training data points. Common distance metrics include:
    1) Euclidean distance
    2) Manhattan distance
    3) Minkowski distance
    
#### 4. Identify Nearest Neighbors:
Sort the distances in ascending order and select the k closest neighbors to the test point.

#### 5. Classify the Test Point:
For classification: Determine the majority class among the k-nearest neighbors and assign that class to the test point.
    
For regression: Calculate the average (or weighted average) of the target variable of the k-nearest neighbors.
    
#### 6. Evaluate the Model:
Use metrics like accuracy, confusion matrix, or mean squared error (for regression) to assess the model's performance on test data.

#### 7. Visualize Results:
Tools like heatmaps and scatter plots are used to display performance metrics and decision boundaries.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
x = df.iloc[:, :-1]  # All rows, all columns except the last one
y = df.iloc[:, -1]   # All rows, only the last column
```

```python
x, y
```

```python
# Assuming x contains the features and y contains the labels
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of K-NN classifier: {accuracy * 100:.2f}%")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="OrRd", xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Pima Dataset')
plt.show()
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of K-NN classifier: {accuracy * 100:.2f}%")
```
